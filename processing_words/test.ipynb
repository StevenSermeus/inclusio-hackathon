```python
import pandas as pd
```

```python
df_ranking = pd.read_csv('../ressources/words_count.csv',sep=',')
```

```python
most_value = -100000
# On parcourt les lignes du dataframe et je veux save dans deux variables le word et le count
for i in range(len(df_ranking)):
    if df_ranking['count'][i] > most_value:
        most_value = df_ranking['count'][i]
        most_name = df_ranking['word'][i]

print(most_name, most_value)
```

```python
dico_word = {}

for i in range(len(df_ranking)):
    dico_word[df_ranking['word'][i]] = (df_ranking['count'][i]/most_value)

print(dico_word)

```

```python
df_ranking_word = pd.DataFrame(dico_word.items(), columns=['word', 'ranking'])

df_ranking_word.to_csv('../ressources/ranking.csv', index=False)
```
